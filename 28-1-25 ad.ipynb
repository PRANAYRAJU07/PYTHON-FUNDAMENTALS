```python
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
from scipy.stats import zscore
data=sns.load_dataset('iris')
df=data.copy()
z_score=np.abs(zscore(df.drop('species',axis=1)))
```

```python
z_score
```

```python
df
```

```python
treshold = 3
non_outliers = (z_score<treshold).all(axis=1)
df_no_outliers=df[non_outliers]
```

```python
df_no_outliers
```

```python
treshold = 3
outliers = (z_score>treshold)
outliers_rows=df[outliers.any(axis=1)]
```

```python
from sklearn.ensemble import IsolationForest
```

```python
data=sns.load_dataset('iris')
data
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
data['species']=lb.fit_transform(data['species'])
```

```python
clf=IsolationForest(random_state=10,contamination=.01)
clf.fit(data)
```

```python
y_pred_outliers = clf.predict(data)
idx=np.where(y_pred_outliers==-1)
idx[0]
```
